###  Reading Operation : read_csv method

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv')

## Display records
df.head(10)
```

###  Reading Operation : read_table method

```python
## Read CSV file 
df=pd.read_table('Dataset/titanic_train1.tsv')

## Display records
df.head()
```

### Both read_csv and read_table can be used interchangeably  

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train1.tsv', sep="\t")

## Display records
df.head()
```

```python
## Read CSV file 
df=pd.read_table('Dataset/titanic_train.csv', sep=",")

## Display records
df.head()
```

## Exploring different parameters in read_csv

### Parameter: header 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv', header=None)

## Display records
df.head()
```

### Parameter: names 

```python
df=pd.read_csv('Dataset/titanic_train.csv', header = 0,
               names=['Column 1','Column 2', 'Column 3', 'Column 4', 'Column 5', 'Column 6','Column 7','Column 8',
                      'Column 9', 'Column 10', 'Column 11', 'Column 12', 'Column 13','Column 14','Column 15'])

## Display records
df.head()
```

### Parameter: index_col 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv', index_col=['passenger_id'])

## Display records
df.head()
```

### Parameter: nrows 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv', nrows=10)

print(df.shape)

## Display records
df.head(20)
```

### Parameter: na_values 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv', na_values=['0'])

## Display records
df.head()
```

### Parameter: skiprows 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train_skiprows_ex.csv',skiprows=[0,1,2])

## Display records
df.head()
```

### Parameter: chunksize 

```python
## Read CSV file 
df=pd.read_csv('Dataset/titanic_train.csv', chunksize=100)

## Display records
df
```
